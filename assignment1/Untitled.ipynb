## numpy.newaxis

```python
a = np.arange(1, 7).reshape(2, 3)
print(a)
print()
print(a[None])
print()
print(a[:, None])
print()
print(a[:, :, None])
```

```python
print(abs(train[None, :] - test[:, None, :]))
```
