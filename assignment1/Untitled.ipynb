```python
import numpy as np
```

```python
train = np.random.rand(4, 5)
test = np.random.rand(2, 5)
print(train)
print()
print(test)
```

```python
train = train.reshape(1, -1)
test = test.reshape(-1, 1)
```

```python
print(train[:5])
print()
print(test[:,:5])
```

```python
print(abs(test - train))
```

```python
x = abs(test - train)[:5,:5]; x
```

```python
np.trace(x)
```

```python
print(np.trace(abs(test - train)))
```

```python
y = abs(test - train).reshape(2, 5, 4, 5); print(y)
```

```python
np.trace(y, axis1=1, axis2=3)
```
